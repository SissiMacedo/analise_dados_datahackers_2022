#4.4 Intervalo de confiança e distribuição amostral

```python
salarios = dados['SALARIO']
salarios
```

```python
#calcular a media dos salarios. Isso nos dá uma estimativa do salário médio na população.
media_amostral = np.mean(salarios)
media_amostral
```

```python
#calcular o desvio padrao da amostra. O desvio padrão nos mostra o quanto os salários variam em relação à média amostral.
desvio_amostral = np.std(salarios)
desvio_amostral
```

```python
#um desvio de 8306 para uma amostra de 9904 é considerado um desvio bem alto, temos uma amostra bem variada
```

```python
nivel_confianca = 0.95
```

```python
#calcular o tamanho da amostra
tamanho_amostra = len(salarios)
tamanho_amostra
```

```python
#calcular o erro padrão. O erro padrão é calculado dividindo o desvio padrão da amostra pela raiz quadrada do tamanho da amostra.
#Ele nos ajuda a entender a precisão da nossa média amostral como uma estimativa da média populacional.
#importar a biblioteca scipy, função stats
from scipy import stats
```

```python
#calcular o erro padrão, usando a função 'sem'
erro_padrao = stats.sem(salarios)
erro_padrao
```

```python
#calcular o intervalo de confiança com a função stats.t.interval, essa função recebe alguns parâmetros que são o nivel de confiança, o numero de graus de
#liberdade da distribuição (tam. amostra - 1), o loc (media amostral), o scale (erro padrao da media).
intervalo_confianca = stats.t.interval(nivel_confianca, tamanho_amostra - 1, loc = media_amostral, scale = erro_padrao)
intervalo_confianca

#Calculamos o intervalo de confiança para um nível de 95% de confiança. Isso significa que, se repetíssemos esse processo muitas vezes,
#95% das vezes o intervalo conteria a verdadeira média populacional.
```

```python
#esse resultado significa que de acordo com nossos dados temos 95% de confiança que a media salarial de pessoas de dados no Brasil é de 9655 reais a
#10153 reais, ou seja, é uma carreira bem promissora. Óbvio que precisamos relembrar que o conceito de media não significa que a maioria receba isso,
#isso seria moda, significa que na tendencia central temos esses valores, mas podemos ter bastante gente ganhando menos e bastante gente ganhando mais.
#Até porque temos diferentes niveis de cargos misturados.
```

###➡️Importância do Intervalo de Confiança:

#####Precisão: O intervalo de confiança nos dá uma faixa dentro da qual acreditamos que a média populacional se encontra, tornando nossa estimativa mais informativa do que uma única média amostral.

#####Incerteza: Reconhece a incerteza inerente à amostragem e nos dá uma ideia de quanto essa incerteza pode ser.

#####Essa aula destacou como podemos usar conceitos estatísticos para fazer estimativas mais confiáveis e informadas sobre uma população, mesmo quando trabalhamos apenas com uma amostra. Entender e calcular intervalos de confiança é uma habilidade crucial na análise de dados, permitindo-nos tomar decisões baseadas em dados com maior certeza e precisão.🧐

####Também calculamos a correlação entre variáveis discretas usando o coeficiente de Cramér.

####Criamos uma função cramer_coeficiente para calcular este coeficiente, utilizando a tabela cruzada gerada com crosstab do pandas e a função chi2_contingency da biblioteca stats.

####Aplicamos essa função para verificar a relação entre cor/raça/etnia e nível de ensino, obtendo um coeficiente de 0.044, indicando quase nenhuma relação. Este resultado sugere que, na nossa amostra, etnias diferentes têm níveis de ensino semelhantes, possivelmente devido ao contexto da área de TI, onde a maioria já tem algum tipo de formação.

```python
#função qui quadrado compara a distribuição observada na tabela cruzada com uma distribuição esperada se as duas variáveis fossem independentes
#uma da outra. Quanto maior a diferença entre a distribuição observada e esperada, maior será o valor do qui quadrado, o que indica uma associação
#mais forte entre as variáveis.
from scipy.stats import chi2_contingency
```

#### Essa função cramer_coeficiente que você criou calcula o Coeficiente V de Cramér (ou Cramér's V), uma métrica muito útil para medir a associação entre duas variáveis categóricas.

O que é o Coeficiente de Cramér?
É uma estatística baseada no teste qui-quadrado que mede a força da associação entre duas variáveis categóricas. Seu valor varia entre 0 (nenhuma associação) e 1 (associação perfeita).

```python
def cramer_coeficiente(coluna1, coluna2):
  tabela_cruzada = np.array(pd.crosstab(coluna1, coluna2))
  chi2 = chi2_contingency(tabela_cruzada)[0]
  soma = np.sum(tabela_cruzada)
  minimo = min(tabela_cruzada.shape) - 1
  cramer = np.sqrt(chi2/(soma*minimo))
  return cramer
```

#### Explicação do código:


Cria uma tabela de contingência (tabela cruzada) entre as duas colunas e transforma em um array NumPy.

Calcula o valor do chi-quadrado usando chi2_contingency do scipy.stats.

Soma total de observações da tabela.

Pega o menor número de categorias entre as duas variáveis (linhas ou colunas), subtrai 1 para o cálculo.

Fórmula do Coeficiente de Cramér.

Retorna o valor entre 0 e 1.




```python
#Mais próximo de 0 menor a relação e mais próximo de 1 maior a relação.
#com esse resultado podemos dizer que as colunas COR  e NIVEL DE ENSINO não tem quase nenhuma relação. Mas, no contexto sabemos que na realidade, cor e
#nivel de ensino são relacionadas na nossa sociedade. Sabemos que pessoas não brancas não tem as mesmas oportunidades de ensino que pessoas brancas têm.
#Porém, porque na nossa análise a correlação foi tão baixa? Muito provavelmente poderíamos dizer que as pessoas que estão na área de TI já tem algum tipo
#de formação, então, mesmo apesar de difentes etnias, a maioria tem uma formação.
cramer_coeficiente(dados['COR/RACA/ETNIA'], dados['NIVEL DE ENSINO'])
```
